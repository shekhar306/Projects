```python
# A general function that returns multiple values
def var_summary(x):
    return pd.Series([x.count(), x.isnull().sum(), x.sum(), x.mean(), x.median(),  x.std(), x.var(), x.min(), x.dropna().quantile(0.01), x.dropna().quantile(0.05),x.dropna().quantile(0.10),x.dropna().quantile(0.25),x.dropna().quantile(0.50),x.dropna().quantile(0.75), x.dropna().quantile(0.90),x.dropna().quantile(0.95), x.dropna().quantile(0.99),x.max()], 
                  index=['N', 'NMISS', 'SUM', 'MEAN','MEDIAN', 'STD', 'VAR', 'MIN', 'P1' , 'P5' ,'P10' ,'P25' ,'P50' ,'P75' ,'P90' ,'P95' ,'P99' ,'MAX'])

num_summary=data_num.apply(lambda x: var_summary(x)).T
```

### Handling Outliers

```python
#Handling Outliers for numerical data - Through function

def outlier_capping(x):
    x = x.clip_upper(x.quantile(0.99))
    x = x.clip_lower(x.quantile(0.01))
    return x

data_num = data_num.apply(lambda x: outlier_capping(x))
```

### Handling Missing Values
#### As we can see in the above audit report some data is missing in numerical data set and some are missing in categorical. We will treat numerical missing data with mean() and categorical data with mode()

```python
#Handling missings

def Num_Missing_imputation(x):
    x = x.fillna(x.mean())
    return x

data_num = data_num.apply(lambda x: Num_Missing_imputation(x))
```

```python
#Handling missings

def Cat_Missing_imputation(x):
    x = x.fillna(x.mode().iloc[0])
    return x

data_cat = data_cat.apply(lambda x: Cat_Missing_imputation(x))
```
