```python
#6. In customer table, create a new variable called “customer value segment” using customer value as follows. -  
#- High Value Segment - > 25000  
#- Medium Value Segment – Between 10000 and 25000  
#- Low Value Segment – less than or equal to 10000 
```

```python
def custvalseg(entry):
    if entry > 25000:
        val = "High Value Segment"
    elif 25000>=entry >10000:
            val="Medium Value Segment"
    else:
                val="Low Value Segment"
                return val
df['customer value segment'] = df['Customer_Value'].apply(custvalseg) 
df
```

```python
#7. Create variables “average revenue per trip” and “balance points” in the customer data set.  
```

```python
df['avg_rev_per-trip'] = df['Customer_Value'] / df['Buy_Times']
df['Balance_points'] = df['Points_Earned'] - df['Points_Redeemed']
df
```

```python
#8. How many days between last purchase date and today?
```

```python
today=pd.datetime.now()
argdate=df.Recent_Date.max()
year = int(argdate / 10000)
month = int((argdate % 10000) / 100)
day = int(argdate % 100)
last_purchase=pd.datetime(year,month,day)
(today-last_purchase).days
```

```python
#9. Calculate percentage of sales by each last city, state and region.
```

```python
total=df.groupby(by=['Last_region'])['Customer_Value'].sum().sum()
((df.groupby(by=['Last_region','Last_state','Last_city'])['Customer_Value'].sum()/total)*100)
```
