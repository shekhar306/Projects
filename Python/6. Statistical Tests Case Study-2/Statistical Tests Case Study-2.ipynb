### 2. An analyst at a department store wants to evaluate a recent credit card promotion. To this end, 500 cardholders were randomly selected. Half received an ad promoting a reduced interest rate on purchases made over the next three months, and half received a standard seasonal ad. Is the promotion effective to increase sales?

```python
Creditpromo=pd.read_csv('creditpromo.csv')
Creditpromo.head(5)
```

```python
Creditpromo.info()
```

```python
# Ho =There is increase in the Sales
# H1 = There is no Increase in the Sales

# Here we are checking for two independent sample therefore,
# we are using Independent Sample t Test.

Standard_dol=Creditpromo.dollars[Creditpromo['insert']=='Standard']
New_Promotion_dol=Creditpromo.dollars[Creditpromo['insert']=='New Promotion']


stats.ttest_ind(a=Standard_dol,b=New_Promotion_dol,equal_var=False) #Assume sample have unequal variance
```

```python
stats.ttest_ind(a=Standard_dol,b=New_Promotion_dol,equal_var=True) #Assume sample have equal variance
```

### 3. An experiment is conducted to study the hybrid seed production of bottle gourd under open field conditions. The main aim of the investigation is to compare natural pollination and hand pollination. The data are collected on 10 randomly selected plants from each of natural pollination and hand pollination. The data are collected on fruit weight (kg), seed yield/plant (g) and seedling length (cm).

```python
Pollination=pd.read_csv('pollination.csv')
Pollination.head(20)
```

```python
Pollination.describe()
```

### Is the overall population of seed yield/plant (g) equals to 200

```python
# Here we are using One Samle T Test
# Ho = Overall population of Seed yield/plant (g) equals to 200
# H1 = Overall population of Seed yield/plant (g) not equals to 200
stats.ttest_1samp(a=Pollination.Seed_Yield_Plant,popmean=200)
```

### Test whether the natural pollination and hand pollination under open field conditions are equally effective or are significantly different.

```python
# H0 = natural pollination and hand pollination under open field condition are equally effective.
# H1=  natural pollination and hand pollination under open field condition are significantly different.
# here i will use Independent Sample t test
natural_poll_Group=Pollination.Fruit_Wt[Pollination['Group']=='Natural']
Hand_poll_Group=Pollination.Fruit_Wt[Pollination['Group']=='Hand']
stats.ttest_ind(a=natural_poll_Group,b=Hand_poll_Group,equal_var=True)
```

```python
stats.ttest_ind(a=natural_poll_Group,b=Hand_poll_Group,equal_var=False)
```
